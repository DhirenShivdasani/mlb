```python
import pandas as pd

# Load the CSV files
prizepicks_data = pd.read_csv('mlb_props.csv')
betting_odds_data = pd.read_csv('test2.csv')

# Display the data to understand its structure
prizepicks_data.head(), betting_odds_data.head()
```

```python
# Merging the datasets on player names and props
merged_data = pd.merge(prizepicks_data, betting_odds_data, left_on='PlayerName', right_on='Name', how='left')

# import ace_tools as tools; tools.display_dataframe_to_user(name="Merged Betting Data", dataframe=merged_data)

# Display the first few rows of the merged data
merged_data.head()
```
